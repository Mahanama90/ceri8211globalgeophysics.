```
import numpy as np
import matplotlib.pyplot as plt
import math
```

```
# Question 03, part(a) Determine North, East & Vertical components
M = 3.87
D = math.radians(62.1)
I = math.radians(59.8)
N = (math.cos(D))*(math.cos(I))*M
E = (math.cos(I))*(math.sin(D))*M
V = (math.sin(I))*M
print('N =', N)
print('E =', E)
print('V =', V)
```

```
# Question 03, part(b)
North = [1.23541566*pow(10,-5),0.910912977*pow(10,-5),0.680789260*pow(10,-5),0.465832032*pow(10,-5),0.259641308*pow(10,-5)]
East  = [2.21051604*pow(10,-5),1.72041523*pow(10,-5),1.23835107*pow(10,-5),0.883529066*pow(10,-5),0.511780975*pow(10,-5)]
# Plot
plt.plot(East,North,'ok')
# Fitting degree 1 polynomial
poly = np.poly1d(np.polyfit(East,North,1)) 
xpoly = np.linspace(0,0.000025,10)

plt.plot(xpoly,poly(xpoly),'-')
plt.xlabel('East Component Am-1')
plt.ylabel('North Component Am-1')
plt.title('Plot of North component vs East component')
plt.grid(True)
plt.show()
slope,intercept = poly
Drad = np.arctan(1/slope)
D    = np.rad2deg(Drad)
print('slope =', slope)
print('Declination =', D)

```
