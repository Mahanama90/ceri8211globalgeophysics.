```
import numpy as np
import matplotlib.pyplot as plt
```

```
# Ratio's of 39Ar/36Ar & 40Ar/36Ar
Ar39 = [1852,1790,1439,3214,2708]
Ar40 = [8855,8439,6867,15380,12970]
# Plot
plt.plot(Ar39,Ar40,'ok')
# Fitting degree 1 polynomial
poly = np.poly1d(np.polyfit(Ar39,Ar40,1)) 
xpoly = np.linspace(1000,4000,10)

plt.plot(xpoly,poly(xpoly),'-')
plt.xlabel('39Ar/36Ar')
plt.ylabel('40Ar/36Ar')
plt.title('Ar-Ar isochron diagram')
plt.grid(True)
plt.show()
slope,intercept = poly

print('slope =', slope)
print('intercept =', intercept)
```
