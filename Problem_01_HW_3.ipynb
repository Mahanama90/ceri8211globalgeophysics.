```
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```
# Parameters
betar = 0.01
gammar = 0.005
er = 0.0001
betaf = 0.005
gammaf = 0.01
ef = 0.0001
# Initial conditions
r0 = 100
f0 = 100
```

```
def func(z,t):
    r,f=z
    return [betar*r-gammar*r-ef*r*f,betaf*f-gammaf*f+er*r*f ]  
```

```
z0=[r0,f0]
t = np.linspace(0,5000,1000)
xx=odeint(func, z0, t)
plt.plot(t, xx[:,0],"-b", label="Rabbits") # Rabbits
plt.plot(t, xx[:,1],"-r", label="Foxes") # Foxes
plt.legend(loc="upper right")
plt.xlabel('Time (Day)')
plt.ylabel('Population')
plt.title('Population Vs Time')
plt.grid(True)
plt.show()
```
