```
import numpy as np
import matplotlib.pyplot as plt
```

```
# Function for polutant concentration
time = np.linspace(0,10000,100)
Conc = (4e4)*time/(10e6+97680*time)
```

```
# Plot 
plt. plot (time,Conc)
plt.xlabel('Time (years)')
plt.ylabel('Polutant concentration (kg/m3)')
plt.title('Concentration of polutnts as a function of time')
plt.grid(True)
plt.show()
```
