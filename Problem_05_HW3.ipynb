```
import numpy as np
import matplotlib.pyplot as plt
```

```
# Ratio's Rb87/Sr86 & Sr87/Sr86
Rb87 = [3.65,1.80,1.48,.82,.66,.74] 
Sr87 = [.7117,.7095,.7092,.7083,.7083,.7082]
# Plot 
plt.plot(Rb87,Sr87,'ok')
# Fitting degree 1 polynomial
poly = np.poly1d(np.polyfit(Rb87,Sr87,1)) 
xpoly = np.linspace(-.5,5,10)
plt.plot(xpoly,poly(xpoly),'-')
plt.xlabel('87Rb/86Sr')
plt.ylabel('87Sr/86Sr')
plt.title('Rb-Sr isotopic curve')
plt.grid(True)
plt.show()
slope,intercept = poly
print('slope =', slope)
print('intercept =', intercept)
```
