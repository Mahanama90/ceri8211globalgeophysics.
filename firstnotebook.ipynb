```
import numpy as np
import matplotlib.pyplot as plt
```

```
le = pow(10,4) # Number of raws of the vector
Nu = le*le # Number of elements in the matrix
x = np.ones((le,1)) # Creating a vector 
A = np.ones(Nu).reshape(le,le) # Creating the matrix by reshaping the elemnts
print('le =', le)
print('Nu =', Nu)
print('x =', x)
print('A =', A)
```

A more efficient way would be to use np.randum.rand(). See its documentation.

```
for p in range(le-1): # fill x from random numbers
 x[p]= np.random.randn()
print('x =', x)
for q in range (le-1): # fill A from random numbers
  for t in range (le-1):
    A[q,t] = np.random.randn()
print ('A =', A)




```

```
b = A.dot(x) # Matrix vector multiplication
print('b =', b)
```

```

plt.scatter(x,b, alpha=0.5, color = 'red', s = 1) # scatter plot of the values of vector b
plt.title("Values of the vector b")
plt.xlabel("x")
plt.ylabel("b")
plt.show()
```
