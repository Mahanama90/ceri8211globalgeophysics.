```
import numpy as np
import matplotlib.pyplot as plt
```

```
# Decay constants 
U235 = 9.85e-10
U238 = 1.55e-10
t = np.linspace(0,5e9,100)
k235 = U235*t
k238 = U238*t
```

```
# Ratio's of 207Pb/235U & 206Pb/238U
xr = np.expm1 (k235) # expm1 gives the exp(k235) - 1 value exactly
yr = np.expm1 (k238) # expm1 gives the exp(k238) - 1 value exactly
plt.plot (xr,yr)
plt.xlabel('207Pb/235U')
plt.ylabel('206Pb/238U')
plt.title('Concordia for U-Pb system')
plt.grid(True)
plt.show()
```

```
zircon_yt = np.array([4233,4201,3780,1713,2595,1177,4356,4235,3674,4126,2604,1961,4364,3355])*1e6 # 206Pb/238U (Myr)
Zircon_xt = np.array([4321,4303,4153,3155,3672,2808,4355,4307,4078,4287,3651,3317,4392,3985])*1e6 # 207Pb/235U (Myr)

# Multipling from decay constants
Zy = U238*zircon_yt
Zx = U235*Zircon_xt
# Calculating the ratio's
Zy_axi = np.expm1(Zy)
Zx_axi = np.expm1(Zx)




```

```
plt.plot (xr,yr,'-')
plt.plot(Zx_axi,Zy_axi,'ok')

# Fitting degree one polynomial towards data points
ply = np.poly1d(np.polyfit(Zx_axi,Zy_axi,1))
xply = np.linspace(0,80,10)
plt.plot (xply, ply(xply), '-')
plt.xlabel('207Pb/235U')
plt.ylabel('206Pb/238U')
plt.title('U-Pb isotope data for a zircon grain')
plt.grid(True)
plt.show()
```
